```python
class ByParagraph:
    
    def __init__(self, text):
        self.text = text
        
    def __iter__(self):
        return self
    
    def __next__(self):
        par =[]
        for line in self.text:
            if line.isspace():
                break
            par.append(line.strip())
        else:
            if not par:
                raise StopIteration()
        return  " ".join(par)
```
