```python
import docx2txt
import spacy
```

```python
my_text = docx2txt.process('resumeD.docx')
print(my_text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sent_token = sent_tokenize(my_text)
```

```python
word_token = word_tokenize(my_text)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
clean_stop_words = []
for rm_w in word_token:
    if rm_w not in stop_words:
        clean_stop_words.append(rm_w)
```

```python
print(clean_stop_words)
```
